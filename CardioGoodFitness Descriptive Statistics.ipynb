### <strong> Calcular Matriz de Correlação com o Método `corr()` </strong>

No código a seguir, está sendo utilizada a **função** `corr()` para calcular a matriz de correlação de um conjunto de dados chamado `mydata`, limitando a análise apenas aos dados numéricos através do **parâmetro** `numeric_only=True`. Em seguida, a matriz de correlação resultante é exibida utilizando a **função** `display()`. A **função** `corr()` pertence à **classe** `DataFrame` da **biblioteca** `pandas`, comumente usada para manipulação e análise de dados em Python.

```python
corr = mydata.corr(numeric_only=True)
display(corr)
```

### <strong> Calcular Matriz de Correlação com o Método `cov()` </strong>

No código a seguir, está sendo calculada a matriz de covariância de um conjunto de dados. A função **`cov`** (Função) é chamada a partir do objeto **`mydata`** (Classe ou Biblioteca, dependendo do contexto), com o parâmetro **`numeric_only=True`** (Parâmetro) para assegurar que apenas colunas numéricas sejam consideradas no cálculo. A covariância é uma medida de como duas variáveis variam juntas. Em seguida, a matriz resultante é exibida utilizando a função **`display`** (Função), que pode fazer parte de uma biblioteca como **IPython** ou **pandas** para mostrar de maneira formatada os dados na saída. Esse código é útil para analisar a relação linear entre diferentes variáveis numéricas presentes no conjunto de dados **`mydata`**.

```python
cov = mydata.cov(numeric_only=True)
display(cov)
```

### <strong> Calcular e Criar Gráfico de Mapa de Calor com a Função `heatmap()` </strong>

No código a seguir, está sendo utilizada a **Biblioteca** `matplotlib` para criar e ajustar uma figura com tamanho personalizado por meio da **Função** `plt.figure` e a **Biblioteca** `seaborn` para gerar um mapa de calor com a **Função** `sns.heatmap`. Esta última recebe vários **Parâmetros**: `annot=True` para mostrar os valores de correlação em cada célula, `cmap='coolwarm'` para definir um esquema de cores que varia de vermelho (positivo) a azul (negativo), `vmin=-1` e `vmax=1` para definir os limites da escala de cores, `center=0` para centralizar a escala de cores em branco, `square=True` para tornar as células quadradas, `linewidths=0.5` para adicionar linhas entre as células, e `fmt='.2f'` para formatar os números com duas casas decimais. A **Função** `plt.title` adiciona um título ao gráfico, e `plt.tight_layout` ajusta o layout para evitar sobreposições. Finalmente, `plt.show` exibe o gráfico na tela.

```python
plt.figure(figsize=(10, 10))

sns.heatmap(corr, annot=True, cmap='coolwarm',  vmin=-1, vmax=1,  center=0,  square=True,  linewidths=0.5,  fmt='.2f') 
plt.title('Mapa de Calor da Matriz de Correlação', fontsize=16, pad=20)
plt.tight_layout()

plt.show()
```

No código a seguir, está sendo utilizado a **biblioteca** `matplotlib` para criar uma figura com dimensões específicas (10x10) e a **biblioteca** `seaborn` para gerar um **mapa de calor** (heatmap) de uma matriz de covariância `cov`. A função **`sns.heatmap`** é configurada com vários **parâmetros**, incluindo `annot=True` para adicionar anotações às células, `cmap='coolwarm'` para definir a paleta de cores, `vmin=-1` e `vmax=1` para definir os limites do mapa de calor, `center=0` para centralizar a coloração, `square=True` para garantir que cada célula seja quadrada, `linewidths=0.5` para definir a largura das linhas de separação entre células e `fmt='.2f'` para formatar os números nas células com duas casas decimais. A função **`plt.title`** adiciona um título ao gráfico com tamanho de fonte 16 e um espaçamento de 20. **`plt.tight_layout`** ajusta automaticamente os parâmetros de preenchimento da figura para que os elementos não se sobreponham. Por fim, **`plt.show`** exibe o gráfico gerado.

```python
plt.figure(figsize=(10, 10))

sns.heatmap(cov, annot=True, cmap='coolwarm',  vmin=-1, vmax=1,  center=0,  square=True,  linewidths=0.5,  fmt='.2f') 
plt.title('Mapa de Calor da Matriz de Covariância', fontsize=16, pad=20)
plt.tight_layout()

plt.show()
```

No código a seguir, está sendo utilizado a **Biblioteca** `sklearn` para realizar uma divisão de dados e aplicar modelos de regressão. Primeiro, os dados são separados em variáveis `X` (com as colunas 'Usage' e 'Fitness') e `y` (com a coluna 'Miles'). Em seguida, a **Função** `train_test_split` divide esses dados em conjuntos de treinamento e teste, utilizando 30% dos dados para teste (`test_size=0.3`) e um estado aleatório fixo (`random_state=0`). Dois modelos diferentes são importados, `LinearRegression` e `LogisticRegression`, ambos **Classes** da biblioteca `sklearn.linear_model`, para possível aplicação nos dados divididos.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression

X = mydata[['Usage', 'Fitness']]
y = mydata[['Miles']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

### <strong> Realizar uma Regressão Linear com a classe `LinearRegression` </strong>

No código a seguir, está sendo realizada uma **Regressão Linear** utilizando a **classe** `LinearRegression` da **biblioteca** `sklearn.linear_model`. Primeiro, é criada uma instância da **classe** `LinearRegression`, armazenada na variável `lr`. Em seguida, o **método** `fit` é chamado para treinar o modelo com os dados de teste `X_test` e `y_test`. Após o treinamento, o **método** `score` é utilizado para calcular a precisão do modelo treinado com base nos mesmos dados de teste `X_test` e `y_test`, retornando um valor que indica a qualidade da predição do modelo sobre esses dados.

```python
lr = LinearRegression() 

lr.fit(X_test,y_test) 
lr.score(X_test,y_test)
```

### <strong> Realizar uma Regressão Logística com a classe `LogisticRegression` </strong>

No código a seguir, está sendo usada a **Biblioteca** `sklearn` para realizar uma regressão logística. Inicialmente, importa-se a **Função** `accuracy_score` da `sklearn.metrics`. Em seguida, cria-se uma instância da **Classe** `LogisticRegression` chamada `model`. O modelo é treinado usando o **Método** `fit` com os dados de treino `X_train` e `y_train`. Depois, realiza-se a predição com o **Método** `predict` nos dados de teste `X_test`, armazenando os resultados em `y_pred`. A precisão das predições é calculada com a **Função** `accuracy_score`, comparando os valores reais `y_test` com as predições `y_pred`, e o resultado é armazenado na variável `k`. Finalmente, a precisão é impressa com a **Função** `print`, exibindo o valor armazenado em `k`.

```python
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
k = accuracy_score(y_test, y_pred)
print('The accuracy is ', k)
```

### <strong> Implementar um classificador de árvore de decisão com a classe `DecisionTreeClassifier` </strong>

No código a seguir, está sendo utilizado a **biblioteca** `sklearn` para implementar um **classificador** de árvore de decisão. Primeiramente, importa-se a **classe** `DecisionTreeClassifier` e o módulo `metrics`. Em seguida, cria-se uma instância de `DecisionTreeClassifier` chamada `dtree`. O método **função** `fit` é usado para treinar o modelo com os dados `X_train` e `y_train`. Após o treinamento, o método `predict` gera previsões para os dados de teste `X_test`. Por fim, a precisão do modelo é calculada utilizando a **função** `accuracy_score` do módulo `metrics`, comparando as previsões (`predictions`) com os valores reais de teste (`y_test`), e o resultado é impresso na tela.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
predictions = dtree.predict(X_test)
print("The accuracy of Decision Tree is:", metrics.accuracy_score(predictions, y_test))
```
